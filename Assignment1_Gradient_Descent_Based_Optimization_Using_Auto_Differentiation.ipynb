(c) Code up the Gradient Decent algorithm from your pseudo-code in actual Python [you should use NumPy expressions] using a suitable starting guess.

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_f(x):
    return 2 * (x - 3.14) * np.exp(-(x - 3.14)**2)

# Gradient Descent function
def gradient_descent(initial_x, learning_rate, num_iterations):
    x = initial_x
    for i in range(num_iterations):
        gradient = gradient_f(x)
        x -= learning_rate * gradient
    return x

# Set initial parameters
initial_guess = 3.0  # Starting guess for x
learning_rate = 0.1
num_iterations = 100

# Run Gradient Descent
optimal_x = gradient_descent(initial_guess, learning_rate, num_iterations)
optimal_y = f(optimal_x)

print("Optimal x:", optimal_x)
print("Optimal y:", optimal_y)
```

**Exercise C.1**

> g(u) = u^4 + 3, u = x-1

> h(x) = x-1


> g'(u) = 4u

> h'(x) = 1

f'(x) = g'(h(x))*h'(x) = 4(x-1)

 **Exercise C.2**

f(x) = 1 + z + z^2 , f(2) = 7

f'(x) = 2z + 1 , f'(2) = 5

```python
import torch
from torch import autograd
import numpy as np
# Implements the function f(z) =1+ z+z^2 using a for loop
def f_loop (z) :
    start =0
    stop_plus_one =3
    step =1
    n_values = np.arange ( start , stop_plus_one , step )
    s =0
    for n in n_values :
        s=s + torch.pow (z ,n )
        n=n +1

    return s

z0 = torch.tensor (2.0, requires_grad = True )
z0
fz0 = f_loop (z0)
fz0
fz0.backward() # calls the autograd engine and makes auto-differentiation
z0.grad
```

```python
import torch
from torch import autograd
def g(w ,y ,z):
   return w**2 + 2*( y**2) + w*y*z

w0 = torch . tensor (1.0 , requires_grad = True )
y0 = torch . tensor (2.0 , requires_grad = True )
z0 = torch . tensor (3.0 , requires_grad = True )

g_value = g(w0 , y0 , z0 ) # print to check = 15.0
g_value
g_value.backward ()

w0.grad # = dfdw (w0 , y0 , z0), should produce tensor (8.)
y0.grad # = dfdy (w0 , y0 , z0), should produce tensor (11.)
z0.grad # = dfdz (w0 , y0 , z0), should produce tensor (2.)
```

**Problem D.1: PyTorch Gradient Descent - one variable**

```python
import math as math
import torch
import matplotlib.pyplot as plt

def f(x):
   return -torch.exp(-(x - torch.tensor(math.pi))**2) + 0.01 * (x - torch.tensor(math.pi))**2

x = torch.linspace(-5, 11, 1000)
y = f(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
import math as math
import torch
from torch import autograd # !!
import matplotlib.pyplot as plt

# Define you objective function
def f(x):
   return -torch.exp(-(x - torch.tensor(math.pi))**2) + 0.01 * (x - torch.tensor(math.pi))**2

tolerance = 1e-6
eta = [1.0,0.1,0.01] # multiplication factor that determines the actual step
# length eta ||g|| , where g is the gradient vector

# Plot the function
plt.figure(figsize=(8, 6))
x = torch.linspace(-5, 11, 1000)
y = f(x)
plt.plot(x,y)

for eta in eta:
  print(eta)
  trajectory_x = []
  trajectory_y = []
  x = torch . tensor ([0.0] , requires_grad = True ) # initial starting guess

  for i in range ( 1000 ):
    # compute function at x
    fx = f(x )

    x_np =x. detach () . numpy ()
    x_np = float ( x_np )
    trajectory_x . append ( x_np )
    f_value_np = fx . detach () . numpy ()
    trajectory_y . append ( f_value_np )

    # compute gradient -> populate  all partial derivatives
    fx.backward ()

    # gradient decent update .
    # ***************************************************
    # *** NOTE SEE TECHNICAL NOTE AT END OF THIS SCRIPT ***
    # ***************************************************
    with torch.no_grad () :
         x -= eta * x.grad # needs to be written like this

    #some_stopping_criteria_fulfilled
    #stopping based on the gradient norm
    if torch.norm(x.grad) < tolerance :
       break

    x.grad.zero_ ()

  # Plot trajectory of each eta
  plt.plot(trajectory_x,trajectory_y,label=f'={eta}')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
